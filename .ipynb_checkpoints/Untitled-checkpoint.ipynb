```python
#disable some annoying warnings
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

#plots the figures in place instead of a new window
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style

style.use('ggplot')
```

```python
crime.head()
```

```python
# Drop null values for district as an amount of null values is low.
crime.dropna(subset=['DISTRICT'], inplace=True)
# Replace to int values
crime['SHOOTING'] = crime['SHOOTING'].apply(lambda s: 0 if str(s) == 'nan' else 1)
```

```python
print (crime.iloc[100])
```

```python
# all crimes with shooting by month and hour
crime_only_shoots = crime.loc[crime['SHOOTING'] == 1] # get all crimes with shooting
plt.scatter(x=crime_only_shoots['HOUR'],y=crime_only_shoots['MONTH'])
plt.xlabel('HOUR')
plt.ylabel('MONTH')
plt.title('HOUR vs MONTH')
plt.show()
```

```python
# all crimes with shooting by month and hour, size and is an amount
crime_only_shoots = crime.loc[crime['SHOOTING'] == 1]
plt.scatter(x=crime_only_shoots['HOUR'],y=crime_only_shoots['MONTH'],s=crime_only_shoots.groupby('HOUR')['SHOOTING'].size())
plt.xlabel('Hour')
plt.ylabel('Month')
plt.title('Hour vs Month, by amount in HOUR')
plt.show()
```

```python
# crimes with shooting per mouth
plt.hist(crime_only_shoots['MONTH'], bins=12)
plt.title("Gaussian Histogram")
plt.xlabel("Moth")
plt.ylabel("Amount")
plt.show()
```

```python
sns.distplot(crime_only_shoots['MONTH'], bins=12)
```

```python
colors = sns.color_palette("hls", 50)
sns.palplot(colors)
```

```python
# group data by DISTRICT and count crimes
count_by_district = crime
count_by_district['count'] = 0

count_by_district = crime.groupby('DISTRICT').agg(['count']).loc[:,['DISTRICT', 'count']]
count_by_district = count_by_district.rename(columns={'count':'count'}).reset_index()
count_by_district.columns = count_by_district.columns.droplevel(1)
count_by_district = count_by_district.sort_values(['count'], ascending=False)

# print(count_by_district)

for (name, group),color in zip(count_by_district.values,colors):
    plt.scatter(x=group,y=group,label=name, c=color,s=30)
plt.yscale('log')
plt.legend()
```

```python
# group data by OFFENSE_CODE and count crimes
count_by_offense_code = crime
count_by_offense_code['count'] = 0

count_by_offense_code = crime.groupby('OFFENSE_CODE').agg(['count']).loc[:,['OFFENSE_CODE', 'count']]
count_by_offense_code = count_by_offense_code.rename(columns={'count':'count'}).reset_index()
count_by_offense_code.columns = count_by_offense_code.columns.droplevel(1)
count_by_offense_code = count_by_district.sort_values(['count'], ascending=False)

# print(count_by_district)

for (name, group),color in zip(count_by_offense_code.values,colors):
    plt.scatter(x=group,y=group,label=name, c=color,s=30)
plt.yscale('log')
plt.legend()
```

```python
###### group data by OFFENSE_CODE_GROUP and count crimes
count_by_offense_code_group = crime
count_by_offense_code_group['count'] = 0

count_by_offense_code_group = crime.groupby('OFFENSE_CODE_GROUP').agg(['count']).loc[:,['count']]
count_by_offense_code_group = count_by_offense_code_group.rename(columns={'count':'count'}).reset_index()
count_by_offense_code_group.columns = count_by_offense_code_group.columns.droplevel(1)
count_by_offense_code_group = count_by_offense_code_group.sort_values(['count'], ascending=False)

objects = count_by_offense_code_group['OFFENSE_CODE_GROUP']
y_pos = np.arange(len(objects))
count = count_by_offense_code_group['count']
ind = np.arange(len(objects))
 
plt.barh(y_pos, count, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Count')
plt.title('Count crimes by offense code')
plt.show()
```

```python
# group data by OFFENSE_CODE_GROUP and count crimes
count_by_offense_code_group_and_shooting = crime.copy()
count_by_offense_code_group_and_shooting['count'] = 0

count_by_offense_code_group_and_shooting = count_by_offense_code_group_and_shooting.groupby('OFFENSE_CODE').agg({'SHOOTING': 'sum', 'count': 'count'}).loc[:,['SHOOTING', 'count']]
count_by_offense_code_group_and_shooting = count_by_offense_code_group_and_shooting.rename(columns={'count':'count', 'SHOOTING': 'shooting_count'}).reset_index()
count_by_offense_code_group_and_shooting.columns = count_by_offense_code_group_and_shooting.columns
count_by_offense_code_group_and_shooting = count_by_offense_code_group_and_shooting.sort_values(['shooting_count'], ascending=False)

print(count_by_offense_code_group_and_shooting)

for (group, shooting_count, count),color in zip(count_by_offense_code_group_and_shooting.values,colors):
    plt.scatter(x=shooting_count,y=count,label=group, c=color,s=150)
plt.yscale('linear')
plt.ylabel('Crimes count')
plt.xlabel('Shooting count')
# plt.legend()
```
