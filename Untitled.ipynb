```python
@interact(bins=(5, 25, 5),color=['red','green','orange','blue'])
def show_distplot(bins,color):
    crime['MONTH'].hist(bins=bins, color=color)
```

```python
@interact(groupBy=['OFFENSE_CODE','OFFENSE_CODE_GROUP','OFFENSE_DESCRIPTION', 'DISTRICT', 'DAY_OF_WEEK'], shooting=[True, False])
def show_distplot(groupBy, shooting):
    count_by_var3 = crime_only_shoots.copy() if shooting else crime.copy()
    count_by_var3['count'] = 0

    count_by_var3 = count_by_var3.groupby(groupBy).agg(['count']).loc[:,['count']]
    count_by_var3 = count_by_var3.rename(columns={'count':'count'}).reset_index()
    count_by_var3.columns = count_by_var3.columns.droplevel(1)
    count_by_var3 = count_by_var3.sort_values(['count'], ascending=False)

    objects = count_by_var3[groupBy]
    y_pos = np.arange(len(objects))
    count = count_by_var3['count']
    ind = np.arange(len(objects))

    plt.barh(y_pos, count, align='center', alpha=0.5)
    plt.yticks(y_pos, objects)
    plt.xlabel('Count')
    plt.title('Count crimes by offense code')
    plt.rcParams["figure.figsize"] = [24,18]
    plt.show()
```

```python
###### group data by YEAR and count crimes
count_by_year_with_avg = crime
count_by_year_with_avg['count'] = 0

count_by_year_with_avg = crime.groupby(['YEAR', 'MONTH']).agg(['count']).loc[:,['count']]
count_by_year_with_avg = count_by_year_with_avg.rename(columns={'count':'count'}).reset_index()
count_by_year_with_avg.columns = count_by_year_with_avg.columns.droplevel(1)
count_by_year_with_avg = count_by_year_with_avg.sort_values(['YEAR', 'MONTH'], ascending=False)
years = count_by_year_with_avg.groupby('YEAR').count().reset_index()['YEAR']
years_amount = np.arange(len(years))

months = count_by_year_with_avg['MONTH'].drop_duplicates()
month_amount = np.arange(len(months))
print(month_amount)

colors = sns.color_palette("hls", 5)

for year, color in zip(years.values, colors):
    plt.plot(count_by_year_with_avg[count_by_year_with_avg['YEAR'] == year]['MONTH'], count_by_year_with_avg[count_by_year_with_avg['YEAR'] == year]['count'], color=color, label='line 1', marker='o', linewidth=2)

average_count_by_month = count_by_year_with_avg.groupby(['MONTH']).mean().reset_index().loc[:,['MONTH', 'count']]
plt.plot(month_amount, average_count_by_month['count'], color='k', label='line 1', marker='o', linewidth=2)
plt.xlabel('Month')
plt.ylabel('Crimes count')
plt.title('Count crimes per month by year')
years[3] = 'avg'
plt.legend(years)
plt.show()
```
