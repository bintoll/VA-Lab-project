```python
# group data by DISTRICT and count crimes
count_by_district = crime
count_by_district['count'] = 0

count_by_district = crime.groupby('DISTRICT').agg(['count']).loc[:,['DISTRICT', 'count']]
count_by_district = count_by_district.rename(columns={'count':'count'}).reset_index()
count_by_district.columns = count_by_district.columns.droplevel(1)
count_by_district = count_by_district.sort_values(['count'], ascending=False)

# print(count_by_district)

for (name, group),color in zip(count_by_district.values,colors):
    plt.scatter(x=group,y=group,label=name, c=color,s=30)
plt.yscale('log')
plt.legend()
```

```python
# group data by OFFENSE_CODE and count crimes
count_by_offense_code = crime
count_by_offense_code['count'] = 0

count_by_offense_code = crime.groupby('OFFENSE_CODE').agg(['count']).loc[:,['OFFENSE_CODE', 'count']]
count_by_offense_code = count_by_offense_code.rename(columns={'count':'count'}).reset_index()
count_by_offense_code.columns = count_by_offense_code.columns.droplevel(1)
count_by_offense_code = count_by_district.sort_values(['count'], ascending=False)

# print(count_by_district)

for (name, group),color in zip(count_by_offense_code.values,colors):
    plt.scatter(x=group,y=group,label=name, c=color,s=30)
plt.yscale('log')
plt.legend()
```

```python
# group data by OFFENSE_CODE_GROUP and count crimes
# there is an error with ascii but it is connected with label - ignore it
count_by_offense_code_group_and_shooting = crime.copy()
count_by_offense_code_group_and_shooting['count'] = 0
count_by_offense_code_group_and_shooting = count_by_offense_code_group_and_shooting.groupby(['OFFENSE_CODE', 'OFFENSE_DESCRIPTION']).agg({'SHOOTING': 'sum', 'count': 'count'}).loc[:,['SHOOTING', 'count']]
count_by_offense_code_group_and_shooting = count_by_offense_code_group_and_shooting.rename(columns={'count':'count', 'SHOOTING': 'shooting_count'}).reset_index()
count_by_offense_code_group_and_shooting.columns = count_by_offense_code_group_and_shooting.columns
count_by_offense_code_group_and_shooting = count_by_offense_code_group_and_shooting.sort_values(['shooting_count'], ascending=False)

for (group, description, shooting_count, count),color in zip(count_by_offense_code_group_and_shooting.values,colors):
    plt.scatter(x=group,y=count, c=color,s=150, marker='o')
    plt.annotate(
        str(description),
        xy=(group, count), xytext=(-20, 20),
        textcoords='offset points', ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
        arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))
plt.yscale('linear')
plt.ylabel('Crimes count')
plt.xlabel('Shooting count')
# plt.legend()

plt.show()
```

```python
from IPython.html.widgets import interact, interact_manual
```

```python
@interact(bins=(5, 25, 5),color=['red','green','orange','blue'])
def show_distplot(bins,color):
    crime['MONTH'].hist(bins=bins, color=color)
```

```python
@interact(groupBy=['OFFENSE_CODE','OFFENSE_CODE_GROUP','OFFENSE_DESCRIPTION', 'DISTRICT', 'DAY_OF_WEEK'], shooting=[True, False])
def show_distplot(groupBy, shooting):
    count_by_var3 = crime_only_shoots.copy() if shooting else crime.copy()
    count_by_var3['count'] = 0

    count_by_var3 = count_by_var3.groupby(groupBy).agg(['count']).loc[:,['count']]
    count_by_var3 = count_by_var3.rename(columns={'count':'count'}).reset_index()
    count_by_var3.columns = count_by_var3.columns.droplevel(1)
    count_by_var3 = count_by_var3.sort_values(['count'], ascending=False)

    objects = count_by_var3[groupBy]
    y_pos = np.arange(len(objects))
    count = count_by_var3['count']
    ind = np.arange(len(objects))

    plt.barh(y_pos, count, align='center', alpha=0.5)
    plt.yticks(y_pos, objects)
    plt.xlabel('Count')
    plt.title('Count crimes by offense code')
    plt.rcParams["figure.figsize"] = [24,18]
    plt.show()
```
